```python
# CCTV 현황 데이터 읽기
import numpy as np
import pandas as pd

# 데이터 로딩
CCTV_Seoul = pd.read_csv(
    './data/01. CCTV_in_Seoul.csv' # 파일의 경로 지정
    , encoding='utf-8'
)
CCTV_Seoul.head(3)
```

```python
# 데이터 탐색
# 정렬, describe, info, head()
CCTV_Seoul.describe()
```

```python
CCTV_Seoul.info()
```

```python
# 정렬해서 데이터 확인 : 소계 ASC
CCTV_Seoul.sort_values(
    by='소계' # 컬럼명
    , ascending=True # default
). head()
```

```python
# 정렬해서 데이터 확인 : 소계 DESC
CCTV_Seoul.sort_values(
    by='소계' 
    , ascending=False
). head()
```

```python
# 최근 증가율 구하고 컬럼 추가
# CCTV_Seoul['최근증가율'] = (14+15+16) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년']\
                        +CCTV_Seoul['2015년']\
                        +CCTV_Seoul['2016년'] ) \
                        / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.head()
```

```python
# 최근 3년간 많이 증가한 구를 확인 : 정렬
CCTV_Seoul.sort_values(
    by='최근증가율'
    , ascending=False
).head()
    
    
```

```python
# 컬럼명 변경 : 기관명 => 구별 변경
CCTV_Seoul.rename(
    columns={
        CCTV_Seoul.columns[0] : '구별'
    }
    , inplace=True
)
CCTV_Seoul.head()
```

```python
# 인구 현황
POP_Seoul = pd.read_excel(
    './data/01. population_in_Seoul.xls'
    , header=2
    , usecols='B, D, G, J, N'
)
POP_Seoul.head()
```

```python
# 컬럼명 변경 : rename()
POP_Seoul.rename(
    columns={
        POP_Seoul.columns[0] : '구별'
        , POP_Seoul.columns[1] : '인구수'
        , POP_Seoul.columns[2] : '한국인'
        , POP_Seoul.columns[3] : '외국인'
        , POP_Seoul.columns[4] : '고령자'
    }
    , inplace=True # 원본에 반영
)
POP_Seoul.head()
```

```python
# 데이터프레임에서 행, 컬럼 삭제 : drop([인덱스], 축axis=0,1, inplace)
# axis = 0 : 행 방향 = 세로?
# axis = 1 : 열 방향 = 가로?
# 합계라는 행 삭제
POP_Seoul.drop(
    [0] # [0] : 0번행 삭제, 리스트 : 여러개 삭제 가능
    , axis=0
    , inplace=True
) # => return : None
# .head() error
```

```python
POP_Seoul.head()
```

```python
# inplace 사용시 주의
# drop(inplace=False) : 복사된 object에서 삭제된 데이터프레임을 반환, header메소드 실행 가능
# drop(inplace=True) : 원본에 반영 = 복사 없다 = None 반환
 
```

```python
# 구별 컬럼에 중복 값 확인 : 중복값 제거 => unique()
POP_Seoul['구별'].unique()
```

```python
# nan 값이 들어있는 행 확인 : 불린 인덱싱 사용
# isnull() : nan 값을 찾아서 boolean return : nan True
POP_Seoul['구별'].isnull()
```

```python
# nan 값이 들어있는 행 추출
POP_Seoul[POP_Seoul['구별'].isnull()]
```

```python
# 인구현황 파일 저장
# to_csv : 데이터프레임을 csv 파일로 저장
POP_Seoul.to_csv('./data/01 인구현황 nan 삭제까지.csv')
CCTV_Seoul.to_csv('./data/01 CCTV 현황 중간 데이터.csv')
```
